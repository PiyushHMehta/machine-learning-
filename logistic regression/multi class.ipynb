```python
%matplotlib inline

from matplotlib import pyplot as plt
from sklearn.datasets import load_digits
```

```python
digits = load_digits()
```

```python
dir(digits)
```

```python
digits.data[0]
```

```python
plt.gray()

for i in range(5):
    plt.matshow(digits.images[i])        
```

```python
digits.target[0: 5]
```

```python
digits.target_names[0: 5]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(digits.data, digits.target, test_size=0.2, random_state=37)
# prevent overfitting
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression(solver='saga', max_iter=5000)
# The convergence error in logistic regression occurs when the optimization algorithm fails to find 
# the optimal coefficients within the specified number of iterations
```

```python
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
plt.matshow(digits.images[37])
```

```python
digits.target[37]
```

```python
model.predict(digits.data[[37]])
```

```python
for i in range(10):
    isCorrectPred = model.predict(digits.data[[i]]) == digits.target[i]
    print(isCorrectPred)
```

```python
plt.matshow(digits.images[5])
```

```python
digits.target[5]
```

```python
model.predict(digits.data[[5]])
```

```python
from sklearn.metrics import confusion_matrix
```

```python
y_predicted = model.predict(X_test)
```

```python
# actual vs predicted
cm = confusion_matrix(y_test, y_predicted)
cm
```

```python
import seaborn as sn

plt.figure(figsize=(10, 7))
sn.heatmap(cm, annot=True)
plt.xlabel('Predicted')
plt.ylabel('Actual')
```
