```python
import pandas as pd
from sklearn.datasets import load_digits
```

```python
digits = load_digits()
```

```python
digits.data.shape
```

```python
dig_0 = digits.data[0].reshape(8, 8)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

plt.gray()
plt.matshow(dig_0)
```

```python
df = pd.DataFrame(digits.data, columns=digits.feature_names)
df.head()
```

```python
X = df
y = digits.target
```

```python
# scaling the features

from sklearn.preprocessing import StandardScaler

scalar = StandardScaler()
X_scale = scalar.fit_transform(X)
X_scale
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_scale, y, test_size=0.2, random_state=37)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_train, y_train)
model.score(X_test, y_test)
```

```python
# pca

from sklearn.decomposition import PCA

pca = PCA(0.95)
X_pca = pca.fit_transform(X)
X_pca.shape

# 95% components remains, 5% removed
# dimesnsion is reduced, cols reducedf rom 64 -> 29(not 29 cols as it is, it calculates new cols)
```

```python
pca.explained_variance_ratio_
```

```python
pca.n_components_
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_pca, y, test_size=0.2, random_state=37)
model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)
model.score(X_test, y_test)
```

```python
pca = PCA(n_components=37)
X_pca = pca.fit_transform(X)
X_train, X_test, y_train, y_test = train_test_split(X_pca, y, test_size=0.2, random_state=37)
model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)
model.score(X_test, y_test)
```

```python
df = pd.read_csv('heart.csv')
df.head()
```

```python
# convert text columns to number
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

le_Sex = le.fit_transform(df.Sex)
le_ChestPainType = le.fit_transform(df.ChestPainType)
le_RestingECG = le.fit_transform(df.RestingECG)
le_ExerciseAngina = le.fit_transform(df.ExerciseAngina)
le_ST_Slope = le.fit_transform(df.ST_Slope)

df['le_Sex'] = le_Sex
df['le_ChestPainType'] = le_ChestPainType
df['le_RestingECG'] = le_RestingECG
df['le_ExerciseAngina'] = le_ExerciseAngina
df['le_ST_Slope'] = le_ST_Slope

df.drop(columns=['Sex', 'ChestPainType', 'RestingECG', 'ExerciseAngina', 'ST_Slope'], inplace=True)

df.head()
```

```python
# z score
from scipy.stats import zscore

z_score = df.apply(zscore)
z_score
```

```python
# remove anything that has Z score > 3 formula or Z score < -3
outliers = (z_score > 3) | (z_score < -3)
outliers_combined = outliers.any(axis=1)
df_cleaned = df[~outliers_combined]
df_cleaned.shape
```

```python
# Apply scaling
from sklearn.preprocessing import StandardScaler
scalar = StandardScaler()

y = df.HeartDisease
df.drop(columns=['HeartDisease'], inplace=True)

scaled_data = scalar.fit_transform(df)
scaled_df = pd.DataFrame(scaled_data, columns=df.columns)
scaled_df.head()
```

```python
scaled_df
```

```python
# Build a classification model using various methods (SVM, logistic regression, random forest)
# and check which model gives you the best accuracy

from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
```

```python
X_train, X_test, y_train, y_test = train_test_split(scaled_df, y, test_size=0.2, random_state=37)
```

```python
svc = SVC()
svc.fit(X_train, y_train)
svc.score(X_test, y_test)
```

```python
lr = LogisticRegression(max_iter=1000)
lr.fit(X_train, y_train)
lr.score(X_test, y_test)
```

```python
rfc = RandomForestClassifier()
rfc.fit(X_train, y_train)
rfc.score(X_test, y_test)
```

```python
# Now use PCA to reduce dimensions, retrain your model and see what impact it has on your model in terms of accuracy. 
# Keep in mind that many times doing PCA reduces the accuracy but computation is much lighter and that's the trade off 
# you need to consider while building models in real life

pca = PCA(n_components=7)
X_pca = pca.fit_transform(scaled_df)
X_train, X_test, y_train, y_test = train_test_split(X_pca, y, test_size=0.2, random_state=37)
model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)
model.score(X_test, y_test)
```

```python
pca = PCA(n_components=7)
X_train, X_test, y_train, y_test = train_test_split(X_pca, y, test_size=0.2, random_state=37)
model = SVC()
model.fit(X_train, y_train)
model.score(X_test, y_test)
```

```python
pca = PCA(n_components=5)
X_train, X_test, y_train, y_test = train_test_split(X_pca, y, test_size=0.2, random_state=37)
model = RandomForestClassifier()
model.fit(X_train, y_train)
model.score(X_test, y_test)
```
