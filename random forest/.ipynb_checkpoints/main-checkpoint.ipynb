```python
import pandas as pd
from sklearn.datasets import load_digits
```

```python
digits = load_digits()
```

```python
X = digits.data
```

```python
y = digits.target
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
plt.gray()
for i in range(5):
    plt.matshow(digits.images[i])
```

```python
df = pd.DataFrame(X)
```

```python
df.head()
```

```python
df['target'] = y
# append new row
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=37)
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model = RandomForestClassifier()
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
model = RandomForestClassifier(n_estimators=50, max_depth=20, min_samples_split=5, min_samples_leaf=4, max_features=6)
# for this pnc, we get the best score
```

```python
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
# confusion matrix
from sklearn.metrics import confusion_matrix
```

```python
y_predicted = model.predict(X_test)
# predicted
```

```python
cm = confusion_matrix(y_test, y_predicted)
cm
```

```python
import seaborn as sn
```

```python
plt.figure(figsize=(10, 7))
sn.heatmap(cm, annot=True)
plt.xlabel('Predicted')
plt.ylabel('Actual')
```

```python
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
dir(iris)
```

```python
X = iris.data
y = iris.target
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=37)
model = RandomForestClassifier()
model.fit(X_train, y_train)
model.score(X_test, y_test)
```

```python
model = RandomForestClassifier(criterion='gini', n_estimators=100)
model.fit(X_train, y_train)
model.score(X_test, y_test)
# gini by def, can use other such as entropy
```
